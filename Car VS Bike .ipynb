### Import Libraries

```python
import numpy as np
import cv2
import os
import random
import matplotlib.pyplot as plt

```

### Load Dataset

```python
DIRECTORY = r'C:\Users\Lenovo\Downloads\archive (18)\Car-Bike-Dataset'
CATEGORIES = ['Bike', 'Car']
```

### Data Preprocessing(Read Images Converted Into Numpy Array Using opencv)

```python
IMG_SIZE = 128
data = []
for category in CATEGORIES:
    folder = os.path.join(DIRECTORY, category)
#     print(folder)
    label = CATEGORIES.index(category)
#     loop through all the images
    for img in os.listdir(folder):
        img_path = os.path.join(folder, img)
#         print(img_path)
#         break
#         convert the images into array
        img_arr = cv2.imread(img_path)
        new_arr = cv2.resize(img_arr, (IMG_SIZE, IMG_SIZE))
        data.append([new_arr, label])
#         plt.imshow(new_arr)
#         break
```

```python
X = []
y = []

for features, label in data:
    X.append(features)
    y.append(label)
```

### Train  Test  Split

```python

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1,random_state=0)
print(len(X_train),len(y_train))
print(len(X_test),len(y_test))

print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
y_test
```

### Build Convolution Neural Network

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
```

```python
model = Sequential()

model.add(Conv2D(64, (3,3), activation = 'relu'))
model.add(MaxPooling2D((2,2)))

model.add(Conv2D(64, (3,3), activation = 'relu'))
model.add(MaxPooling2D((2,2)))

model.add(Flatten())

model.add(Dense(128, input_shape = X.shape[1:], activation = 'relu'))

model.add(Dense(1, activation = 'sigmoid'))
```

```python
# model.summary()
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=5, validation_split=0.1)
```

```python
X_train.shape
```

```python
model.evaluate(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix , classification_report
y_pred = model.predict(X_test)
y_pred_classes = [0 if element < 0.5 else 1 for element in y_pred]

print("Classification Report: \n", classification_report(y_test, y_pred_classes))
```
